```python
import time
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

def prod_non_zero_diag_vec(x):
    x = np.diagonal(x)
    x = x[x != 0]
    x = np.multiply.reduce(x)
    return x

def prod_non_zero_diag(x):
    sum = 1
    for i in range(0, len(x)):
        for j in range(0, len(x[i])):
            if (i == j):
                if (x[i][j] != 0):
                    sum *= x[i][j]
    return sum

data_size = []
time_non_vectorized = []
time_vectorized = []
time_scipy_func = []
for i in range(20):
    x = np.random.rand(200 * i, 200 * i)
    data_size.append(4000 * i * i)
    R1 = time.time()
    prod_non_zero_diag(x)
    time_non_vectorized.append((time.time() - R1) * 4000)
    R2 = time.time()
    prod_non_zero_diag_vec(x)
    time_vectorized.append((time.time() - R2) * 4000) 


    
fig = plt.figure(figsize=(13, 10))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax2.plot(data_size, time_non_vectorized, color='red', marker = 'o', label=u"Non vectorized")
ax2.plot(data_size, time_vectorized, marker = 'x', label=u"Vectorized",)
ax2.set_title(u"prod_non_zero_diag", fontsize = 18)
ax2.set_xlabel(u"Matrix size", fontsize = 18)
ax2.set_ylabel(u"Time (s)", fontsize = 18)
ax2.grid()
ax2.legend(loc = 'best', fontsize = 20)

fig.show()  
```

```python
def are_multisets_equal(x, y):
    x.sort()
    y.sort()
    fl = True
    for i in range(len(x)):
        if (x[i] != y[i]):
            fl = False
    return fl

def are_multisets_equal_vec(x, y):
    x = np.sort(x)
    y = np.sort(y)
    fl = False
    fl = np.array_equal(x, y)
    return fl
    
data_size = []
time_non_vectorized = []
time_vectorized = []
for i in range(1, 30):
    x = np.random.rand(i*100)
    y = np.random.rand(i*100)
    data_size.append(i * 100)
    R1 = time.time()
    are_multisets_equal(x, y)
    time_non_vectorized.append((time.time() - R1) * 1000)
    R2 = time.time() * 1000
    are_multisets_equal_vec(x, y)
    time_vectorized.append((time.time() - R2) * 1000)

fig = plt.figure(figsize=(13, 10))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax2.plot(data_size, time_non_vectorized, color='red', marker = 'o', label=u"Non vectorized")
ax2.plot(data_size, time_vectorized, marker = 'x', label=u"Vectorized",)
ax2.set_title(u"are_multisets_equal", fontsize = 18)
ax2.set_xlabel(u"Size of the largest array", fontsize = 18)
ax2.set_ylabel(u"Time (s)", fontsize = 18)
ax2.grid()
ax2.legend(loc = 'best', fontsize = 20)

fig.show()
```

** 1.3 Найти максимальный элемент в векторе среди элементов, перед которыми стоит нулевой**

```python
def max_after_zero(x):
    a = []
    for i in range(0, len(x) - 1):
        if (x[i] == 0):
            a.append(x[i + 1])
    if (len(a) != 0):
        return max(a)
    else:
        return 0

def max_after_zero_vec(x):
    zero_vector = (x == 0)
    max_vector = x[1:][zero_vector[:-1]]
    if (len(max_vector) != 0):
        return max_vector.max()
    else:
        return 0
    
data_size = []
time_non_vectorized = []
time_vectorized = []
for i in range(20):
    x = np.array([k % 10 for k in range(i * 200)])
    data_size.append(200 * i)
    R1 = time.time()
    max_after_zero(x)
    time_non_vectorized.append((time.time() - R1) * 4000)
    R2 = time.time()
    max_after_zero_vec(x)
    time_vectorized.append(time.time() - R2) * 4000)
   
fig = plt.figure(figsize=(13, 10))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax2.plot(data_size, time_non_vectorized, color='red', marker = 'o', label=u"Non vectorized")
ax2.plot(data_size, time_vectorized, marker = 'x', label=u"Vectorized",)
ax2.set_title(u"max_after_zero", fontsize = 18)
ax2.set_xlabel(u"Size of the array", fontsize = 18)
ax2.set_ylabel(u"Time (s)", fontsize = 18)
ax2.grid()
ax2.legend(loc = 'best', fontsize = 20)

fig.show()
```

```python
import time
import matplotlib.pyplot as plt
import numpy as np
from scipy.spatial import distance


%matplotlib inline

def pairwise_distance(x, y):
    from math import sqrt
    D = np.zeros((x.shape[0], y.shape[0]), dtype=np.float)

    for i in range(x.shape[0]):
        s = 0.0
        for j in range(y.shape[0]):
            s = (x[i] - y[j]) ** 2
            s = np.sum(s)
            s = sqrt(s)
            D[i][j] = s
    return D


def pairwise_distance_vec(x, y):
    from math import sqrt
    D = []
    for i in range(len(x)):
        D.append([0.0] * len(y))
        for j in range(len(y)):
            sum = 0.0
            for k in range(len(x[0])):
                sum += (x[i][k] - y[j][k]) ** 2
            sum = sqrt(sum)
            D[i][j] = sum

    return D

def Euclid(x, y):
    return distance.cdist(x, y, 'euclidean')
    
data_size = []
time_non_vectorized = []
time_vectorized = []
time_euclid = []
for i in range(10):
    x = np.random.rand(i*100, i * 10)
    y = np.random.rand(i * 100, i * 10)
    data_size.append(1000 * i * i)
    R1 = time.time()
    pairwise_distance_vec(x, y)
    time_non_vectorized.append((time.time() - R1) * 1000)
    R2 = time.time()
    pairwise_distance_vec(x, y)
    time_vectorized.append((time.time() - R2) * 1000)
    t = time.time()
    Euclid(x, y)
    time_euclid.append((time.time()- t) * 1000)
    
fig = plt.figure(figsize=(13, 8))
ax2 = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax2.plot(data_size, time_non_vectorized, color='red', marker = 'o', label=u"Non vectorized")
ax2.plot(data_size, time_vectorized, marker = 'x', label=u"Vectorized",)
ax2.plot(data_size, time_euclid, marker = 'o', label=u"Scipy")
ax2.set_title(u"Pairwise distance", fontsize = 15)
ax2.set_xlabel(u"Data_size", fontsize = 15)
ax2.set_ylabel(u"Time (ms)", fontsize = 15)
ax2.grid()
ax2.legend(loc = 'best', fontsize = 20)

fig.show()
```

### Задача 3: полные данные
__(1 балл)__

Воспользуйтесь файлами attendance.csv и school_support.txt для того, чтобы добавить новые признаки в данные. Желательно по максимуму использовать возможности pandas для упрощения преобразований.

Решение : attendance - это таблица из 32 столбцов, в каждом из котором стоит либо +, либо ничего. Пробегаем по attendanceи меняем + на 1, а пробелы на 0, после чего складываем и получаем уровень посещяемости по шкале от 0 до 32.

school_support - состоит из списка с номерами учащихся (это можно было понять, так как там записаны номера от 0 до 649). Добавляем в общие данные колонку school_support и ставим 1 на против студента, если его номер входил в файл school_support. 


```python
attendance = pd.read_csv('/Users/valerystoeva/Downloads/homework-practice-01-3/attendance.csv', sep=';')

for i in attendance:
    attendance.loc[attendance[i].isnull(), i] = 0 #первое по строке, второй по столбцу< 1 раз - изменение в 1 столбце
    attendance.loc[attendance[i] == '+', i] = 1

attendance = attendance.apply(np.sum, axis=1)#каждому элементу столбца присвоит значение в кавычках
attendance = attendance.rename('attendance')

pass_data = pass_data.join(attendance)

ss = pd.read_csv('/Users/valerystoeva/Downloads/homework-practice-01-3/school_support.txt', header=None)
school_support = pd.DataFrame(data=np.zeros(pass_data.shape[0]), columns=['school_support'], dtype=int)

ss = ss.loc[:, 0].values # лок - по всем строкам взяли элемент из первого столбца и избавились от двухмерности, val - превращает в список
school_support.loc[ss] = 1

pass_data = pass_data.join(school_support)

print(pass_data.head(5))
```

Найдите в данных сломанный признак (он не соответствует описанию) и исправьте его.

Решение : Смотрим на данные. Последный столбец явно неисправен. Он состоит из "склеившихся" столбцов plans_university и past_failures. В plans_university должны хранится 1 и -1, а в past_failures данные по некой шкале. Оставим в plans_university только первую цифру и знак просто поделив на 10 без остатка. А в past_failures запишем вторую цифру. 

```python
# добавили новый столбец в данные c правильными значениями

past_failures = pass_data['plans_universitypast_failures'].apply(lambda x : abs(x) % 10)
past_failures = past_failures.rename('past_failures')
pass_data = pass_data.join(past_failures)

#переименовали правильно исходный столбец
pass_data = pass_data.rename(index=str, columns={'plans_universitypast_failures': 'plans_university'})
pass_data['plans_university'] = pass_data['plans_university'].apply(lambda x : int(x / 10))

print(pass_data.head(5))
```

### Задача 5: борьба с выбросами
__(1.5 балла)__

Качество предсказания может ухудшаться, если в данных присутствуют корректные значения признаков (с точки зрения чтения данных и применения методов), но не соответствующие реальным объектам. Например, данные могли быть введены в неверном формате, а потом слишком грубо приведены к общему виду, из-за чего ошибка не была замечена.
Попробуем от такого избавиться &mdash; а для этого такие объекты нужно сначала найти. Конечно, нам еще недоступны многие продвинутые способы, но давайте попробуем обойтись простыми.

Первый способ это сделать &mdash; посмотреть для каждого признака на распределение его значений и проверить крайние значения на правдоподобность.

*Hint 1: используйте функцию DataFrame.hist*

*Hint 2: в описании датасета выше есть информация, необходимая для восстановления правильных значений*

Решение : Выводим гистограммы для каждого столбца. Некоторые из них не соответствуют описанию. Например,в параметре age в некотрых строчках записан не возраст, а год рождения. Меняем эти строки, и заполняем возрастом студента в 2006 году.  После чего, выводим гистогаамму и видим, что все получилось, топерь она соответствует описанию признака. 

```python
pass_data['age'].plot.hist() #before
plt.show()
```

```python
pass_data.loc[pass_data['age'] > 50, 'age'] = pass_data.loc[pass_data['age'] > 50, 'age'].apply(lambda x: 2006 - x)
pass_data['age'].plot.hist() #after
plt.show()
```

В traveltime тоже есть ошибка. Вмето условной шкалы от 0 до 4, вводили время поездки в минутах. Но это просто исправить, заменив минуты на значения по шкале.

```python
pass_data['traveltime'].plot.hist() #before
plt.show()
```

```python
def time(x):
    if (x < 15):
        return 1
    if (x >= 15 and x < 30):
        return 2
    if (x >= 30 and x < 60):
        return 3
    if (x > 60):
        return 4

pass_data.loc[pass_data['traveltime'] > 4, 'traveltime'] = pass_data.loc[pass_data['traveltime'] > 4, 'traveltime'].apply(lambda x: time(x))
pass_data['traveltime'].plot.hist() #after
plt.show()
```
